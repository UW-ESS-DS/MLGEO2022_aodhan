```python
# Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
```

# Create a list and write a function

```python
# First we create list
list_item = [1,2,3,4,5]

# Next we create a function
def func_multiplier(list_item, multiplication_factor):
    """
    This function multiplies a given list by a factor input by user.
    
    Inputs:
    -------
    list_item: input list (type: list)
    multiplication_factor: factor by which to multiply list (int or float)
    
    Returns:
    --------
    multiplied_array: array version of list_item after multiplied by multiplication_factor (numpy array)
    """
    new_array = np.array(list_item) # Declare list as array
    multiplied_array = new_array*multiplication_factor # multiply array by factor
    
    # Now return the product array
    return(multiplied_array)
array_item = func_multiplier(list_item, 2)
print(array_item)    
```

```python
# create an array of random numbers
time_index = range(0,100)
random_y_values = np.random.rand(100)
```

```python
# Plot random values as dependent variables versus time_index
plt.plot(time_index, random_y_values)
plt.title('Random Values Timeseries')
plt.ylabel('Random Value')
plt.xlabel('Time')
plt.grid()
plt.savefig('RandomValueTimeseries.png', dpi=300)
```
